```python
import pandas as pd
import numpy as np
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = "generated_data/purchase_data_3.csv"
data_df = pd.read_csv(csv_path)
data_df.head()
```

```python
#Player Count
#Counting the number of unique players, since some might have made multiple purchases
player_count = len(data_df.groupby("SN"))
player_count_df = pd.DataFrame({"Total Players": [player_count]})
#player_count_df = data_df.groupby("SN").count()
player_count_df.head()
```

```python
#Purchasing Analysis (Total)
#Number of Unique Items
unique_items = len(data_df.groupby('Item Name'))

#Average Purchase Price
average_price = data_df['Price'].mean()

#Total Number of Purchases
purchase_count = len(data_df.groupby('Purchase ID'))

#Total Revenue
total_revenue = data_df['Price'].sum()

#Create data frame with purchasing info
purchase_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], 
                                     "Average Price": [average_price], 
                                     "Number of Purchases": [purchase_count], 
                                     "Total Revenue": [total_revenue]})
#Still haven't figured out how to do the $ sign for price and revenue
#purchase_analysis_df['Average Price'] = purchase_analysis_df['Average Price'].apply(lambda x: '${:,.2f}MM'.format(x))
purchase_analysis_df
```

```python
#Gender Demographics
#Percentage and Count of (Unique) Male Players
male_df = data_df.loc[data_df["Gender"] == "Male"]
male_count = len(male_df.groupby("SN"))
male_count
#Percentage and Count of (Unique) Female Players
female_df = data_df.loc[data_df["Gender"] == "Female"]
female_count = len(female_df.groupby("SN"))
female_count
#Percentage and Count of (Unique) Other / Non-Disclosed
other_df = data_df.loc[data_df["Gender"] == "Other / Non-Disclosed"]
other_count = len(other_df.groupby("SN"))

#Percentages
male_percentage = (male_count / player_count) * 100
female_percentage = (female_count / player_count) * 100
other_percentage = (other_count / player_count) * 100

#Creating data frame with column and row labels
gender_analysis_df = pd.DataFrame(index=("Male","Female","Other"),
                                  columns=("Percentage of Players", 
                                     "Total Count"))

#Populating data frame
gender_analysis_df.loc['Male'] = [male_percentage, male_count]
gender_analysis_df.loc['Female'] = [female_percentage, female_count]
gender_analysis_df.loc['Other'] = [other_percentage, other_count]

gender_analysis_df
```

```python
#Purchasing Analysis (Gender)
#The below each broken by gender
#Purchase Count
male_purchase_count = len(male_df)
female_purchase_count = len(female_df)
other_purchase_count = purchase_count - male_purchase_count - female_purchase_count

#Average Purchase Price
male_average_price = male_df['Price'].mean()
female_average_price = female_df['Price'].mean()
other_average_price = other_df['Price'].mean()

#male_average_price = male_df['Price'].mean()
#Total Purchase Value
total_male_revenue = male_df['Price'].sum()
total_female_revenue = female_df['Price'].sum()
total_other_revenue = other_df['Price'].sum()
#Normalized Totals
#I'm assuming the normalized totals remove the highest and lowest prices 
#since I'm not sure what the definition of normalized value is
male_normalized_avgprice = (total_male_revenue - male_df['Price'].min() - male_df['Price'].max()) / (len(male_df)  - 2)
female_normalized_avgprice = (total_female_revenue - female_df['Price'].min() - female_df['Price'].max()) / (len(female_df)  - 2)
#Knowing that there is only one other purchase, how do I calculate the normalized total? 
other_normalized_avgprice = other_average_price

#Creating the data frame for displaying purposes
purchase_analysis_gender_df = pd.DataFrame(index=("Male","Female","Other"),
                                  columns=("Purchase Count", 
                                           "Average Purchase Price",
                                     "Total Purchase Value",
                                          "Normalized Totals"))

#Populating data frame
purchase_analysis_gender_df.loc['Male'] = [male_purchase_count, male_average_price, 
                                           total_male_revenue, male_normalized_avgprice ]
purchase_analysis_gender_df.loc['Female'] = [female_purchase_count, female_average_price, 
                                           total_female_revenue, female_normalized_avgprice ]
purchase_analysis_gender_df.loc['Other'] = [other_purchase_count, other_average_price, 
                                           total_other_revenue, other_normalized_avgprice ]

purchase_analysis_gender_df
```

```python
#Age Demographics:
#- Establish the age bins: "<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"
age_group1_df = data_df.loc[data_df["Age"] < 10]
age_group2_df = data_df.loc[(data_df["Age"] >= 10) & (data_df["Age"] <= 14)]
age_group3_df = data_df.loc[(data_df["Age"] >= 15) & (data_df["Age"] <= 19)]
age_group4_df = data_df.loc[(data_df["Age"] >= 20) & (data_df["Age"] <= 24)]
age_group5_df = data_df.loc[(data_df["Age"] >= 25) & (data_df["Age"] <= 29)]
age_group6_df = data_df.loc[(data_df["Age"] >= 30) & (data_df["Age"] <= 34)]
age_group7_df = data_df.loc[(data_df["Age"] >= 35) & (data_df["Age"] <= 39)]
age_group8_df = data_df.loc[data_df["Age"] > 40]
 
#- Categorize the existing players using the age bins

#- Calculate the Numbers and Percentages by Age Group

age1_count = len(age_group1_df.groupby("SN"))
age2_count = len(age_group2_df.groupby("SN"))
age3_count = len(age_group3_df.groupby("SN"))
age4_count = len(age_group4_df.groupby("SN"))
age5_count = len(age_group5_df.groupby("SN"))
age6_count = len(age_group6_df.groupby("SN"))
age7_count = len(age_group7_df.groupby("SN"))
age8_count = len(age_group8_df.groupby("SN"))

age1_percentage = (age1_count / player_count) * 100
age2_percentage = (age2_count / player_count) * 100
age3_percentage = (age3_count / player_count) * 100
age4_percentage = (age4_count / player_count) * 100
age5_percentage = (age5_count / player_count) * 100
age6_percentage = (age6_count / player_count) * 100
age7_percentage = (age7_count / player_count) * 100
age8_percentage = (age8_count / player_count) * 100
#- Display Age Demographics Table
age_analysis_df = pd.DataFrame(index=("<10", "10-14", "15-19", "20-24", "25-29", 
                                      "30-34", "35-39", "40+"),
                                  columns=("Percentage of Players", 
                                           "Total Count"))

age_analysis_df.loc['<10'] = [ age1_count, age1_percentage ]
age_analysis_df.loc['10-14'] = [age2_count, age2_percentage]
age_analysis_df.loc['15-19'] = [age3_count, age3_percentage]
age_analysis_df.loc['20-24'] = [age4_count, age4_percentage]
age_analysis_df.loc['25-29'] = [age5_count, age5_percentage]
age_analysis_df.loc['30-34'] = [age6_count, age6_percentage]
age_analysis_df.loc['35-39'] = [age7_count, age7_percentage]
age_analysis_df.loc['40+'] = [age8_count, age8_percentage]
age_analysis_df

```
